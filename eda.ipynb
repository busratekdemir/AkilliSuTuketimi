Yıllara Göre Ortalama Tüketim

```python
yillik = df.groupby("YIL")["TUKETIM"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(data=yillik, x="YIL", y="TUKETIM", marker="o")
plt.title("Yıllara Göre Ortalama Tüketim")
plt.show()
```

Aylık Sezonsallık Analizi

```python
aylik = df.groupby("AY")["TUKETIM"].mean().reset_index()

plt.figure(figsize=(8,5))
sns.lineplot(data=aylik, x="AY", y="TUKETIM", marker="o")
plt.title("Aylık Ortalama Tüketim (Sezonsallık)")
plt.show()
```

İlçe Bazlı Ortalama Tüketim

```python
ilce_avg = df.groupby("ILCE")["TUKETIM"].mean().reset_index().sort_values("TUKETIM")

plt.figure(figsize=(12,8))
sns.barplot(data=ilce_avg, x="TUKETIM", y="ILCE")
plt.title("İlçe Bazlı Ortalama Tüketim")
plt.show()
```

Korelasyon Matrisi

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="Blues", fmt=".2f")
plt.title("Korelasyon Matrisi")
plt.show()
```

# SmartWater – Veri Ön İşleme ve Hazırlık

Bu defterde veri birleştirme, eksik değer analizi, aykırı değer/ anomali temizliği ve model için hazır veri setinin oluşturulması yapılacaktır.


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/master_3kaynak_2020_2024.csv")
df.head()
```

Veri Setinin Yapısına Bakma

```python
df.info()
df.describe().T
```

Eksik Değer Analizi

```python
df.isnull().sum()
```

Eksik Değer Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Eksik Değer Görselleştirme")
plt.show()
```

Aykırı Değer Temizliği

```python
df = df[df["TUKETIM"] > 0]   # 0 veya negatif tüketim olamaz
df = df[df["YIL"] >= 2020]   # sistem yılları 2020-2024
df = df[df["YIL"] <= 2024]
```

Anomali TemizliğiAnomali TemizliğiAnomali Temizliği

```python
df = df[df["IS_ANOMALY"] == 0]
```

Kullanılacak Kolonları Seçme

```python
feature_cols = [
    "YIL", "AY",
    "ORTALAMA_TUKETIM_MEAN",
    "ABONE_ADEDI_SUM",
    "KAYIT_SAYISI",
    "ORT_DOLULUK",
    "TOPLAM_SU",
    "BARAJ_SAYISI"
]

target_col = "TUKETIM"

df = df[feature_cols + [target_col]]
df.head()
```

Eksik Değer Doldurma

```python
df[feature_cols] = df[feature_cols].fillna(df[feature_cols].median(numeric_only=True))
df.isnull().sum()
```

```python
df.describe().T
```

Temiz Veri Setini Kaydetme

```python
df.to_csv("data/temiz_su_tuketimi_ml_ready.csv", index=False)
print("Temiz veri kaydedildi → temiz_su_tuketimi_ml_ready.csv")
```
